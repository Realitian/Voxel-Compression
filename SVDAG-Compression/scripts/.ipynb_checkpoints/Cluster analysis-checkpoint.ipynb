```python
import community # louvain clustering
import markov_clustering as mcl
# import infomap
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
%matplotlib notebook

color_keys = list(mcolors.TABLEAU_COLORS.keys())
```

```python
def draw_clusters(G, partition):
    size = float(len(set(partition.values())))
    pos = nx.kamada_kawai_layout(G)
#     pos = nx.spectral_layout(G)
#     pos = nx.spring_layout(G)

    count = 0.
    for com in set(partition.values()) :
        count = count + 1.
        list_nodes = [nodes for nodes in partition.keys()
                                    if partition[nodes] == com]

        nx.draw_networkx_nodes(G, pos, list_nodes, node_size = 40,
                                    node_color = mcolors.TABLEAU_COLORS[color_keys[int(count) % len(color_keys)]])
        
    nx.draw_networkx_edges(G, pos, alpha=0.5)
    plt.axis('off')
```

```python
file_in = '../bin/edges12-nice-one.txt'
edges = []
with open(file_in) as f:
    for line in f:
        n1, n2, w = line.split(' ')
        edges.append((int(n1), int(n2), float(w)))
        # edges.append((int(n2), int(n1), float(w)))

G = nx.Graph()

G.add_weighted_edges_from(edges)

# G.add_edge(0, 1, weight=1)
# G.add_edge(1, 2, weight=1)
# G.add_edge(0, 2, weight=1)
# G.add_edge(3, 2, weight=1)
# G.add_edge(3, 4, weight=1)
# G.add_edge(0, 4, weight=1)
# G.add_edge(2, 4, weight=1)
# G.add_edge(0,  5, weight=1)

G_int = nx.convert_node_labels_to_integers(G)

numrows = 2
numcols = 5

### compute clusters (partitions)
fig, axes = plt.subplots(nrows=2, ncols=5)
fig.patch.set_visible(False)
# plt.tight_layout()

# Markov clustering
inflations = [1.2, 1.5, 2, 4, 6]
for i, p in enumerate(inflations):
    matrix = nx.to_numpy_matrix(G_int)
    mcl_res = mcl.run_mcl(matrix, inflation=p)
    clusters = mcl.get_clusters(mcl_res)
    partition = {}
    for c_num, cluster in enumerate(clusters):
        for node in cluster:
            partition[node] = c_num
    
    plt.subplot(numrows, numcols, i+1)
    if i == 2:
        plt.title('Markov Clustering')
    plt.text(1, 0.7, '{}'.format(p), horizontalalignment='right')
    if i == 0:
        plt.text(1, 0.9, 'Inflation:', horizontalalignment='right')
    draw_clusters(G_int, partition)
    
# Louvain modularity
resolutions = [4, 1, 0.5, 0.1, 0.05] # similar to markov
resolutions = [4, 2, 1, 0.5, 0.1]
for i, p in enumerate(resolutions):
    partition = community.best_partition(G_int, resolution=p)
    
    plt.subplot(numrows, numcols, i + numcols + 1)
    if i == 2:
        plt.title('Louvain Clustering')
    plt.text(1, 0.7, '{}'.format(p), horizontalalignment='right')
    if i == 0:
        plt.text(1, 0.9, 'Resolution:', horizontalalignment='right')
    draw_clusters(G_int, partition)


plt.show()
plt.savefig("mcl-vs-louvain.svg")
```
